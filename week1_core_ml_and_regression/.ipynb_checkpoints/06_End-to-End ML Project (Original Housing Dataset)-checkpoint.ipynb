**Dataset:** Original Housing dataset (housing.csv)  
**Goal:** Predict median house value using all features.  
This notebook includes:
- Full dataset inspection
- Exploratory Data Analysis (EDA)
- Feature engineering
- Stratified sampling
- Transformation pipelines
- Multiple regression models
- GridSearch for hyperparameter tuning
- Model evaluation and comparison
- Model persistence (saving/loading)


```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#model selection and preprocessing
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit , cross_val_score , GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
#machine learnign models
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
#evaluation 
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#saving and loading the model
import joblib
```

```python
#loading the dataset
housing = pd.read_csv("housing.csv")
housing.head()
```

```python
#inspect the data
housing.info()
```

```python
#statistics
housing.describe()
```

```python
#value counts for categorical feature
housing["ocean_proximity"].value_counts()
```

```python
#checking for missing values
housing.isnull().sum()
```

```python
#stratified sampling based on median_income
housing['income_cat'] = pd.cut(
    housing['median_income'],
    bins = [0,1.5,3.0,4.5, 6., np.inf],
    labels = [1,2,3,4,5]
)
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size = 0.2, random_state = 42)
```

```python
for train_index, test_index in split.split(housing,housing['income_cat']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
for set_ in (strat_train_set,strat_test_set):
    set_.drop('income_cat', axis=1, inplace = True)
```

```python
#Separate Features and Labels
housing_train = strat_train_set.copy()
housing_labels = housing_train['median_house_value'].copy()
housing_train = housing_train.drop('median_house_value', axis=1)
```

```python
#EDA / Visualization
housing_train.hist(bins=50, figsize=(12,8))
plt.show()
# Use original dataset for correlation, not the train split
corr_matrix = housing.select_dtypes(include=['int64', 'float64']).corr()

print(corr_matrix['median_house_value'].sort_values(ascending=False))
```

```python
#Preparing Pipelines
num_features = housing_train.select_dtypes(include=[np.number]).columns.tolist()
cat_features = ['ocean_proximity']

# Numerical pipeline
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical pipeline
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())
])

# Full pipeline
full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])
```

```python
#prepraing training data
housing_prep = full_pipeline.fit_transform(housing_train)
```

```python
#Training Models
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(housing_prep, housing_labels)

# Decision Tree
tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(housing_prep, housing_labels)

# Random Forest
forest_reg = RandomForestRegressor(random_state=42)
forest_reg.fit(housing_prep, housing_labels)
```

```python
#evaluating on training data
def display_scores(model, X, y):
    predictions = model.predict(X)
    mse = mean_squared_error(y, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, predictions)
    print(f"RMSE: {rmse}")
    print(f"R2: {r2}\n")

print("Linear Regression:")
display_scores(lin_reg, housing_prep, housing_labels)

print("Decision Tree Regression:")
display_scores(tree_reg, housing_prep, housing_labels)

print("Random Forest Regression:")
display_scores(forest_reg, housing_prep, housing_labels)
```

```python
#cross validation
forest_scores = cross_val_score(forest_reg, housing_prep, housing_labels, scoring='r2', cv=5)
print("Random Forest CV R2 scores:", forest_scores)
print("Mean CV R2:", forest_scores.mean())
```

```python
#grid search
param_grid = [
    {'n_estimators': [50, 100], 'max_features': [6, 8]},
    {'bootstrap': [False], 'n_estimators': [50, 100], 'max_features': [6, 8]}
]

grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='r2', return_train_score=True)
grid_search.fit(housing_prep, housing_labels)
grid_search.best_params_
```
